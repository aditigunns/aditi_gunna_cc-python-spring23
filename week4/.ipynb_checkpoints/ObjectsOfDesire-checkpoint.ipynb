Create a class object of something that you love. Write at least 3 methods for it. Include some attributes.

Items you love can be categorical: 

* Food  
* Pets 
* Vehicles 
* Places 
* Activities 
* Things 


```python
class HumanBody:
    
    #super class with basic human attributes
    
    def __init__(self, limbs, fingers, eyeColor, age, name):
        self.limbs = limbs #number of limbs the human body has
        self.fingers = fingers #number of fingers the human body has
        self.eyeColor = eyeColor 
        self.age = age
        self.name = name
        
    def display(self):
        '''Displays the attributes of the human being'''
        print(f'{self.name} is a {self.age} year old human being with {self.limbs} limbs, {self.fingers} fingers, and their eyes are {self.eyeColor} colored.')
        
    
class FemaleBody(HumanBody):
    
    #class with the female body attributes and functions, inherits the HumanBody class
    
    def __init__(self, limbs, fingers, eyeColor, age, name, height, weight, braSize, hairLength, shoeSize, apparelSize, mgID):
        self.liberalized = False
        super().__init__(limbs, fingers, eyeColor, age, name) #basic human attributes of the female body
        self.height = height #height of the female body
        self.weight = weight #weight of the female body
        self.braSize = braSize #bra size of the female body
        self.hairLength = hairLength #length of the female body's hair
        self.shoeSize = shoeSize #shoe size of the female body
        self.apparelSize = apparelSize #clothing sizes of the female body
        self.mgID = mgID #Non-name identifier for the female body
        
    def display(self):
        '''Displays the attributes of the female body, if liberated then also displays the human attributes of the 
        female body. 
        '''
        if(self.liberalized):
            print(f'{self.name} is liberated.')
            super().display()
        print(f'''
        Female's ID Number: {self.mgID}
        Height: {self.height} cms
        Weight: {self.weight} kgs
        Bra Size: {self.braSize}
        Hair Length: {self.hairLength}
        Shoe Size: {self.shoeSize}
        Apparel Size: {self.apparelSize}
        ''')
        
            
    def Check(self):
        '''Compares the Female body to the ideal standards of contemporary female objectification and displays appropriate
        responses to the current state of existence
        '''
        print(f"Female's ID Number: {self.mgID}")
        #using references from here(https://www.businessinsider.com/miss-universe-winners-body-types-2016-4)
        
        if(self.height < 168):
            print(f"Height: {self.height} cms is not tall enough")
        elif(self.height > 178):
             print(f"Height: {self.height} cms is too tall")
        else:
            print(f"Height: {self.height} cms is perfect")
                
        if((self.weight/((self.height/100)**2)) < 16):
            print(f"Weight: {self.weight} kgs is too thin")
        elif((self.weight/((self.height/100)**2)) > 18):
             print(f"Height: {self.weight} kgs is too fat")
        else:
             print(f"Weight: {self.weight} kgs is perfect")
                
        print(f"Bra Size: {self.braSize} (shouldn't they be bigger?)")
        
        print(f"Hair Length: {self.hairLength} (hmmmmmmmmmmmmmmmmm)")
        
        if(self.shoeSize < 5):
            print(f"Shoe Size: {self.shoeSize} is too small")
        elif(self.shoeSize > 8):
             print(f"Shoe Size: {self.shoeSize} is too big")
        else:
             print(f"Shoe Size: {self.shoeSize} cms is perfect")
                
        if(self.apparelSize > 8):
            print(f"Apparel Size: {self.apparelSize} is too big")
        else:
            print(f"Apparel Size: {self.apparelSize} is perfect")
            
            
        
    def liberate(self):
        '''The female body is now liberated, and no longer objectified. They identify as a human being, not an object
        '''
        self.liberalized = True
        self.height = 'Tall Enough'
        self.weight = 'A happy and healthy amount'
        self.braSize = 'None of your business'
        self.hairLength = 'Who even cares'
        self.shoeSize = 'Who wants know'
        self.apparelSize = 'Nope, just no'
        self.mgID = 'Learn my name'
        
```

```python
aditi = FemaleBody(4,10,'Brown',24,'Aditi',178,75,'34B','BelowShoulder',10,14,'12we4rt')
```

```python
aditi.display()
aditi.Check()
aditi.liberate()
print('\n')
aditi.display()
```
